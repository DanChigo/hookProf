```python
import torch
import torch.nn as nn
import time
import numpy as np
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear1 = nn.Linear(10000, 4000)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(4000, 1000)

    def forward(self, x):
        x = self.linear1(x)
        x = self.relu(x)
        x = self.linear2(x)
        return x

control_model = Model().to(device)  # No hooks
hooked_model = Model().to(device)   # Both hooks

x = torch.randn(6400, 10000).to(device)
y = torch.randn(6400, 1000).to(device)

iterations = 100

time_control_list = []
time_hooked_list = []

# Hook functions
def forward_hook(module, input, output):
    _ = output.mean().item()

def backward_hook(module, grad_input, grad_output):
    _ = grad_output[0].norm().item()

# Attach forward + backward hooks to hooked_model
hook_handles = []
for module in hooked_model.modules():
    if isinstance(module, nn.Linear):
        hook_handles.append(module.register_forward_hook(forward_hook))
        hook_handles.append(module.register_backward_hook(backward_hook))

# Profiling loop
for i in range(iterations):
    # Control Model (No Hooks)
    optimizer_control = torch.optim.SGD(control_model.parameters(), lr=0.01)
    criterion_control = nn.MSELoss()

    torch.cuda.synchronize()
    start_time = time.time()

    output = control_model(x)
    loss = criterion_control(output, y)

    optimizer_control.zero_grad()
    loss.backward()
    optimizer_control.step()

    torch.cuda.synchronize()
    end_time = time.time()

    time_control_list.append(end_time - start_time)

    # Hooked Model (Both Hooks)
    optimizer_hooked = torch.optim.SGD(hooked_model.parameters(), lr=0.01)
    criterion_hooked = nn.MSELoss()

    torch.cuda.synchronize()
    start_time = time.time()

    output = hooked_model(x)
    loss = criterion_hooked(output, y)

    optimizer_hooked.zero_grad()
    loss.backward()
    optimizer_hooked.step()

    torch.cuda.synchronize()
    end_time = time.time()

    time_hooked_list.append(end_time - start_time)

# Calculate means
mean_control = np.mean(time_control_list)
mean_hooked = np.mean(time_hooked_list)

# Save mean times to separate files
with open('control.txt', 'a') as f:
    f.write(f"{mean_control:.6f}s\n")

with open('hook.txt', 'a') as f:
    f.write(f"{mean_hooked:.6f}s\n")

# Plotting results
plt.figure(figsize=(10, 6))
plt.plot(range(iterations), time_control_list, label='Control (No Hooks)', marker='o')
plt.plot(range(iterations), time_hooked_list, label='Hooked (Forward + Backward Hooks)', marker='x')
plt.xlabel('Iteration')
plt.ylabel('Execution Time (seconds)')
plt.title('Profiling Results: Control vs Hooked Model')
plt.legend()
plt.grid(True)
plt.savefig('profiling_results_plot.png')
plt.show()
```

```python
# Read the control times
with open('control.txt', 'r') as f:
    control_times = np.array([float(line.strip().replace('s', '')) for line in f])
# Read the hook times
with open('hook.txt', 'r') as f:
     hook_times = np.array([float(line.strip().replace('s', '')) for line in f])

# Sanity check: Make sure the lengths match
if len(control_times) != len(hook_times):
    raise ValueError("The number of entries in control.txt and hooks.txt must be the same.")

# Calculate percent differences
percent_differences = ((hook_times - control_times) / control_times) * 100

plt.figure(figsize=(10, 6))
bars = plt.bar(range(1, len(percent_differences) + 1), percent_differences, color='royalblue')

# Add values on top of each bar
for i, bar in enumerate(bars):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{percent_differences[i]:.2f}%', ha='center', va='bottom')
plt.xlabel('Run Number')
plt.ylabel('Percent Difference (%)')
plt.title('Percent Difference Between Hooked and Control Execution Times (Bar Graph)')

# Save plot
plt.savefig('percent_difference_bar_plot.png')

# Show plot
plt.show()

# Optionally: Save the percent differences to a file if you want
np.savetxt('percent_differences.txt', percent_differences, fmt='%.4f')
print(np.mean(percent_differences))
```

## Takeaways 
A conventional SGD model w forward and backwards hooks is around 2.5% slower. This is a negligble overhead, telling us that hooks are not a good place to target for optimizations. Additionally, this overhead decreases as one increases the tensor sizes. Increasing the second parameter in the second linear layer from 100 to 1000 drops the percent difference to around 1.75% which can be seen in the iterations after 10.
